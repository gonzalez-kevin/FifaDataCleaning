# Cleaning Data

## Check Club Column

## Check Contract Column

```python
# should be string
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique() # inconsistent formatting
```

```python
# look for rows with different formatting
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
# function to specify start date, end date, and contract length given contract not Free or On loan

def extract_contract_info(contract):
    """function to specify start date, end date, and contract length given contract not Free or On loan"""
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
    return start_date, end_date, contract_length

# apply function to Contract column and create new columns
new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

# insert new cols after Contract column
for i in range(len(new_cols)):
    fifa.insert(loc = fifa.columns.get_loc('Contract') + 1 + i, column = new_cols[i], value = new_data[i])
```

```python
fifa.sample(3)
```

```python
def contract_category(contract):
    """function witll categorize contract status"""
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'
    
# Add status category column
fifa.insert(loc = fifa.columns.get_loc('Contract Length(years)') + 1, column = 'Contract Status', value = fifa['Contract'].apply(contract_category))
```

```python
fifa.sample(3)
```

## Check Height Column

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
def height_conversion(height):
    """function to standardize height in cm"""
    if 'cm' in height:
        return int(height.strip(' cm'))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet) * 12 + int(inches.strip('"'))
        return round(total_inches * 2.54)
    
# apply height conversion fn to height column
fifa['Height'] = fifa['Height'].apply(height_conversion)
```

```python
fifa['Height'].unique()
```

```python
fifa = fifa.rename(columns = {'Height': 'Height(cm)'})
fifa.sample(1)
```

## Check Weight Column

```python
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

```python
def weight_conversion(weight):
    """function to standardize weight in kg"""
    if 'kg' in weight:
        return int(weight.strip('kg'))
    else:
        lbs = int(weight.strip('lbs'))
        return round(lbs/2.205)
    
# apply weight conversion to weight column
fifa['Weight'] = fifa['Weight'].apply(weight_conversion)
```

```python
fifa['Weight'].unique()
```

```python
fifa = fifa.rename(columns = {'Weight': 'Weight(kg)'})
```

```python
fifa.sample(1)
```
